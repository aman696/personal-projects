```python
stemmer = LancasterStemmer()
def clean_text(text):
    text = text.lower()                                 
    text = re.sub('-',' ',text.lower())   
    text = re.sub(f'[{string.digits}]',' ',text) 
    text = ' '.join([stemmer.stem(word) for word in text.split()])  
    text =  re.sub(r'@\S+', '',text)                    
    text =  re.sub(r'http\S+', '',text)                  
    text =  re.sub(r'pic.\S+', '',text) 
    text =  re.sub(r"[^a-zA-Z+']", ' ',text)             
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text+' ')      
    text = "".join([i for i in text if i not in string.punctuation])
    words = nltk.tokenize.word_tokenize(text,language="english", preserve_line=True)
    text = " ".join([i for i in words if len(i)>2])
    text= re.sub("\s[\s]+", " ",text).strip()           
    return re.sub(f'[{re.escape(string.punctuation)}]','',text) 
train_data['Cleaned_Description'] = train_data.Description.apply(clean_text)
test_data['Cleaned_Description'] = test_data.Description.apply(clean_text)

train_data.head()
num_words = 50000
max_len = 250
tokenizer = Tokenizer(
    num_words=num_words,
    filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~',
    lower=True)
text_to_tokenize = train_data['Cleaned_Description']
tokenizer.fit_on_texts(text_to_tokenize)
```
