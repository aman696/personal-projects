```python
import cv2
import numpy as np
import matplotlib.pyplot as mp
try:
    image=r"C:\Users\Aman Chaturvedi\Downloads\WhatsApp Image 2023-08-26 at 16.56.08.jpeg"
    img = cv2.imread(image)#i was getting an error thats why needed to use this
    if img is None:
        raise ValueError("Failed to load image. Check the image path.")
except Exception as e:
    print("Error:", e)
    exit()
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
mp.imshow(img_rgb)#needed to convert the image coloring
mp.axis('off') 
mp.show()

face= cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')#loads pretrained ai face detection models
face2 = face.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30), flags=cv2.CASCADE_SCALE_IMAGE)
for (x, y, w, h) in face2:
    cv2.rectangle(img_rgb, (x, y), (x + w, y + h), (255, 0, 0), 2)
mp.imshow(img_rgb)
mp.axis('off')
mp.show() 
```

```python
import cv2#video detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
video = cv2.VideoCapture(0)
def detect_bounding_box(frame):
    frame1= cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(frame1, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30), flags=cv2.CASCADE_SCALE_IMAGE)
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)  
    return frame
while True:
    ret, faces = video.read()  
    if not ret:
        break
    frame= detect_bounding_box(faces)
    cv2.imshow('Face Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('a'):
        break
video.release()
cv2.destroyAllWindows()
```
